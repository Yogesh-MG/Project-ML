### SEA Level Predictor

```python
#get Modules 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as se
from scipy.stats import linregress as lin
#get data
df = pd.read_csv('sea.csv')
x=df['Year']
#get copy 
df1 = df.copy()
df1 = df1[df1['Year']>1999]
year = np.arange(2000,2051)
#get graph
plt.figure(figsize=(10,5))
plt.style.use(['dark_background'])
plt.scatter(x=df['Year'],y=df['CSIRO Adjusted Sea Level'],color='r',edgecolors='w',label='CASL')
reg = lin(x=df['Year'],y=df['CSIRO Adjusted Sea Level'])
reg1 = lin(x=df1['Year'],y=df1['CSIRO Adjusted Sea Level'] )
plt.plot(df['Year'],reg.intercept+reg.slope*x ,color='darkblue',label='fitted line',linestyle='-')
plt.plot(year,reg1.intercept+reg1.slope*year ,color='orange',label='fitted line')
le = reg1.intercept+reg1.slope*2050
plt.plot([2050,2050],[0,le],color='green',linestyle='--')
plt.xlabel('Year')
plt.ylabel('CISRO Adjusted Sea level')
plt.legend()
le1= reg1.intercept+reg1.slope*2050
plt.show()
#get predicted sea level
print(f'\nThe CSIRO adjusted prediction in 2050 will be {le1}')
```
